# Warning inplace parameter

inplace parameter overwrite the current index!!!

```
data8 = pd.read_csv('data.txt')

data8.columns = ["ID", "Address", "City", "ZIP", "Country", "Name", "Employees"]

data8
```

```
data8.set_index("ID", inplace=True)

data8
```

```
data8.set_index("Address", inplace=True) #The ID will be overwriten

data8
```

```
data8.set_index("Name", inplace=True, drop=False) #To not drop index column from data frame

data8
```
