# K-Means Clustering

Data: https://drive.google.com/file/d/1HcxYRfJGM95swdJFWKo2WHgJLf2cfHZ5/view?usp=sharing

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the elbow method to find the optimal number of clusters

```
from sklearn.cluster import KMeans
wcss = []
models = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    models.append(kmeans)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method', fontdict={'color':'white'})
plt.xlabel('Number of clusters', fontdict={'color':'white'})
plt.ylabel('WCSS', fontdict={'color':'white'})
plt.tick_params(colors='w')
plt.show()
```

## Training the K-Means model on the dataset

```
kmeans = models[4]
y_kmeans = kmeans.fit_predict(X)
```

## Visualising the clusters

```
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers', fontdict={'color':'white'})
plt.xlabel('Annual Income (k$)', fontdict={'color':'white'})
plt.ylabel('Spending Score (1-100)', fontdict={'color':'white'})
plt.legend()
plt.tick_params(colors='w')
plt.show()
```
